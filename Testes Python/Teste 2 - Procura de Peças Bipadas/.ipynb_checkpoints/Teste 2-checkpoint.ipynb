## Função de Procura de Peças

```python
def procuraPeças(list1, list2):
    
    arrayResult = []
    
    for item in list1:
        if item in list2:
            arrayResult.append([item, 1, "Peça Encontrada no Sistema"])
        else:
            arrayResult.append([item, 0, "Peça não Encontrada no Sistema"])
    
    data = np.array(arrayResult)
    
    return pd.DataFrame(data, columns=["Peças", "Valor Status", "Descrição Status"])
```

## Função de Comparação - Peças com e sem Final

```python
def comparaPeçasComESemFinal(list1, list2):
    
    arrayResult = []
    
    for item in list1:
        for j in list2:
            if str(item) in str(j):
                arrayResult.append([item, j])
    
    data = np.array(arrayResult)
    
    return pd.DataFrame(data, columns=["Peça - Código sem Local", "Peça - Código com Local"])
```

## Função de Conversão - DataFrame to List

```python
def dataFrameToList(df):
    pecas = []
    
    for item in df.values:
        pecas.append(item[0])
    
    return pecas
```

## Função de Transformação - DataFrame em Excel

```python
def dataFrameToExcel(df, endereco):
    df.to_excel(endereco, index=False)
```

## Funções de Transformação - Local Coletado para Local VA

```python
def defineRua(string):
    
    if string == '1':
        return 'A'
    elif string == '2':
        return 'B'
    elif string == '3':
        return 'C'
    elif string == '4':
        return 'D'
    elif string == '5':
        return 'E'
    elif string == '6':
        return 'F'
    else:
        return "ERRO"

def defineColunaEGaveta(n):
    
    col = str(n // 10)
    gav = str(n % 10)
    
    if(len(str(gav)) == 1):
        return " G{} C0{}".format(col, gav)
    else:
        return " G{} C{}".format(col, gav)

def transformaLocal(df, coluna):
    
    df[coluna] = df[coluna].map(lambda x: "R" + defineRua(str(x)[0]) + defineColunaEGaveta(int(str(x)[1:4])))
```

## Função de Transformação - Formato coleta para formato Virtual Age

```python
def coletaParaVA(df):
    df.insert(1, "Local", ['' for i in range(0, len(df.index))], allow_duplicates = True)
    
    df["Local"] = df["Peças"].map(lambda x: int(str(x)[-6:]))
    transformaLocal(df, "Local")
    df["Peças"] = df["Peças"].map(lambda x: int(str(x)[:-6]))

    return df
```

## Execução do Teste

```python
dfColetaComLocal = coletaParaVA(dfColetaComFinal)
dfColetaComLocal
```

```python
result = pd.merge(dfVAComLocal, dfColetaComLocal, how="inner", on="Peças")
result
```

```python
listaPecasVASemFinal = dataFrameToList(dfVASemLocal)
listaPecasColetaSemFinal = dataFrameToList(dfColetaSemFinal)
listaPecasColetaComFinal = dataFrameToList(dfColetaComFinal)
```

```python
dfFinal1 = procuraPeças(listaPecasColetaSemFinal, listaPecasVASemFinal)
dfFinal1
```

```python
dfFinal2 = comparaPeçasComESemFinal(dataFrameToList(dfFinal1[dfFinal1["Valor Status"] == '0']), listaPecasColetaComFinal)
dfFinal2
```

```python
dataFrameToExcel(dfFinal1[dfFinal1["Valor Status"] == '0'], "output/Resultados da Comparação.xlsx")
dataFrameToExcel(dfFinal2, "output/Equivalência de Peças Com E Sem Local.xlsx")
```
